```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,2*np.pi,100) #创建x轴数据
y = np.sin(x) #创建y轴数据

fig, ax = plt.subplots() #创建一个包含一个子图的画布和轴对象
ax.plot(x,y) # 在当前子图上绘制折线图

plt.show() #显示图形
```

```python
x=[1,2,3,4]
y=[1,4,2,3]
fig,ax=plt.subplots()
ax.plot(x,y)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
plt.show()
```
