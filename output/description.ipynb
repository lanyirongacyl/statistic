```python
import pandas as pd  
import matplotlib.pyplot as plt  
  
# 假设你的DataFrame是df  
x = df['民族'].value_counts().index  
y = df['民族'].value_counts(normalize=True).values * 100  
  
fig, ax2 = plt.subplots()  
ax2.bar(x, y, color='blue')  # 可以调整颜色为其他你喜欢的颜色  
  
labels = df['民族'].value_counts().index  
ax2.set_xticks(x)  # 设置x轴刻度  
ax2.set_xticklabels(labels)  # 设置x轴刻度标签  
ax2.set_ylabel('百分比')  # 设置y轴标签  
ax2.set_title('各民族占比')  # 设置标题  
  
plt.show()
```

```python
from pyreadstat import pyreadstat
#df2可以重命名为国家认同调查数据表
国家认同调查数据表,matedata = pyreadstat.read_sav(R'C:\Users\LENOVO\Desktop\statistic\data\identity.sav',apply_value_formats=True,formats_as_ordered_category=True) 
```

```python
国家认同调查数据表['年级'].value_counts(sort=False) #ddf2=国家认同调查数据表
```

```python
result = 国家认同调查数据表['年级'].value_counts(sort=False)   #索引不叫values,为什么叫一列（序列），因为index不是数值
df_result = pd.DataFrame(result)               #DataFtame是创建二维表格，.类似于“的”
df_result['累计求和'] = df_result['count'].cumsum()  #cumsum累计的结果，result.values.cumsum()（原代码）
# df_result['新变量1'] = 1 #False是可以任意进行数值赋值的，对此进行变量
# df_result['新变量2'] = df_result.新变量1 + 1
df_result['比例'] = 国家认同调查数据表['年级'].value_counts(normalize=True)
df_result['累计比例'] = df_result['比例'].cumsum
```

```python
x = df_result.index
y = df_result['count'].values
fig,ax2 = plt.subplots()
ax2.scatter(x,y)
plt.show()
```
